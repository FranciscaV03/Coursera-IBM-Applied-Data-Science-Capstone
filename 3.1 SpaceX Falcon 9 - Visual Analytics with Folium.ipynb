```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# 1) Put your candidate points here (one or many per category).
#    If you only have one per category, keep the list length = 1.
proximities = {
    "City": [
        ("Cape Canaveral", 28.405840, -80.604771),
        # ("Cocoa Beach", 28.320006, -80.607551),
    ],
    "Railway": [
        ("KSC Industrial Rail", 28.572000, -80.667000),
    ],
    "Highway": [
        ("NASA Causeway", 28.520000, -80.676000),
    ],
    "Coastline": [
        ("Coastline point", 28.563670, -80.571630),
    ],
}

# 2) Helper: find nearest point within a category
def nearest_point(site_lat, site_lon, points):
    best = None  # (name, lat, lon, distance_km)
    for name, plat, plon in points:
        d = calculate_distance(site_lat, site_lon, plat, plon)
        if (best is None) or (d < best[3]):
            best = (name, plat, plon, d)
    return best

# 3) Add a marker for the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    tooltip="Launch Site",
    icon=folium.Icon(color="blue", icon="flag"),
).add_to(site_map)

# 4) For each category, place the nearest marker, label distance, and draw a line
results = []  # optional: collect distances for a table later
for category, candidates in proximities.items():
    name, plat, plon, dist_km = nearest_point(launch_site_lat, launch_site_lon, candidates)

    # marker at the nearest point
    folium.Marker(
        [plat, plon],
        tooltip=f"{category}: {name}",
        icon=folium.Icon(color="gray", icon="info-sign"),
    ).add_to(site_map)

    # distance label at that point
    folium.Marker(
        [plat, plon],
        icon=DivIcon(
            icon_size=(180, 24),
            icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;color:#d35400;"><b>{dist_km:.2f} KM</b></div>',
        ),
    ).add_to(site_map)

    # line from launch site to the nearest point
    line = folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [plat, plon]],
        weight=2,
        tooltip=f"{category}: {name} — {dist_km:.2f} km",
    )
    site_map.add_child(line)

    results.append((category, name, round(dist_km, 3)))

# (optional) quick printout
for c, n, d in results:
    print(f"{c}: closest is {n} at {d} km")

site_map  # display the updated map
```
